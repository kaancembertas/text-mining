# DATA MINING PROJECT

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk.download('punkt')
nltk.download('stopwords')
```

### PREPROCESSING FUNCTIONS

```python
def remove_stopwords (tokenized_text):
    stop_words = set(stopwords.words("english"))
    new_data = [x for x in tokenized_text if x not in stop_words]
    return new_data
```

```python
def apply_stemming (tokenized_text):
    stemmed_words = []
    ps = PorterStemmer()
    for word in tokenized_text:
        new_word = ps.stem(word)
        stemmed_words.append(new_word)
    return stemmed_words
```

```python
def remove_single_characters (tokenized_text):
    new_data = [word for word in tokenized_text if len(word)>1]
    return new_data
```

```python
def remove_by_filters (tokenized_text,filters=[]):
        new_data = [x for x in tokenized_text if x not in filters]
        return new_data
```

### READING DOCUMENTS

```python
real_positive_path = Path('opinion_spam/positive/truthful')
fake_positive_path = Path('opinion_spam/positive/deceptive')
real_negative_path = Path('opinion_spam/negative/truthful')
fake_negative_path = Path('opinion_spam/negative/deceptive')

def read_documents(path):
    files = path.iterdir()
    document = ""

    for index,file in enumerate(files, start=0):
        reader = open(file,"r")
        text = reader.read().lower()
        document += " "+text
    return document

real_positive_doc = read_documents(real_positive_path)
fake_positive_doc = read_documents(fake_positive_path)
real_negative_doc = read_documents(real_negative_path)
fake_negative_doc = read_documents(fake_negative_path)
```

### Tokenization and Preprocessing
- Real Positive

```python
token_real_positive = word_tokenize(real_positive_doc)
token_real_positive = remove_stopwords(token_real_positive)
token_real_positive = apply_stemming(token_real_positive)
token_real_positive = remove_single_characters(token_real_positive)
```

- Fake Positive

```python
token_fake_positive = word_tokenize(fake_positive_doc)
token_fake_positive = remove_stopwords(token_fake_positive)
token_fake_positive = apply_stemming(token_fake_positive)
token_fake_positive = remove_single_characters(token_fake_positive)
```

- Real Negative

```python
token_real_negative = word_tokenize(real_negative_doc)
token_real_negative = remove_stopwords(token_real_negative)
token_real_negative = apply_stemming(token_real_negative)
token_real_negative = remove_single_characters(token_real_negative)
```

- Fake Negative

```python
token_fake_negative = word_tokenize(fake_negative_doc)
token_fake_negative = remove_stopwords(token_fake_negative)
token_fake_negative = apply_stemming(token_fake_negative)
token_fake_negative = remove_single_characters(token_fake_negative)
```

#### Explore Real Positive Data

```python
real_positive_freq_dist = FreqDist(token_real_positive)
real_positive_freq_dist.plot(30,cumulative=False)
plt.show()
print("Number of Features: "+ str(len(real_positive_freq_dist)))
real_positive_freq_dist.most_common(15)
```

#### Explore Fake Positive Data

```python
fake_positive_freq_dist = FreqDist(token_fake_positive)
fake_positive_freq_dist.plot(30,cumulative=False)
plt.show()
print("Number of Features: "+ str(len(fake_positive_freq_dist)))
fake_positive_freq_dist.most_common(15)
```

#### Explore Real Negative Data

```python
real_negative_freq_dist = FreqDist(token_real_negative)
real_negative_freq_dist.plot(30,cumulative=False)
plt.show()
print("Number of Features: "+ str(len(real_negative_freq_dist)))
real_negative_freq_dist.most_common(15)
```

#### Explore Fake Negative Data

```python
fake_negative_freq_dist = FreqDist(token_fake_negative)
fake_negative_freq_dist.plot(30,cumulative=False)
plt.show()
print("Number of Features: "+ str(len(fake_negative_freq_dist)))
fake_negative_freq_dist.most_common(15)
```

### Extracting non-unique words

```python
filters = ['room','hotel','stay']
token_real_positive = remove_by_filters(token_real_positive,filters)
token_fake_positive = remove_by_filters(token_fake_positive,filters)
token_real_negative = remove_by_filters(token_real_negative,filters)
token_fake_negative = remove_by_filters(token_fake_negative,filters)

real_positive_freq_dist = FreqDist(token_real_positive)
fake_positive_freq_dist = FreqDist(token_fake_positive)
real_negative_freq_dist = FreqDist(token_real_negative)
fake_negative_freq_dist = FreqDist(token_fake_negative)
```

```python
real_positive_freq_dist.plot(30, cumulative=False, title="Real Positive")
plt.show()

fake_positive_freq_dist.plot(30, cumulative=False, title="Fake Positive")
plt.show()

real_negative_freq_dist.plot(30, cumulative=False, title="Real Negative")
plt.show()

fake_negative_freq_dist.plot(30, cumulative=False, title="Fake Negative")
plt.show()
```
